### Sintasis de Loc es
### [ row_initial:row_final, col_initial:col_final]
Nombres de las columnas y los renglones

```python
dF.loc[3:10,:'Glasses']
```

### Sintasis de iLoc es
### [ row_initial:row_final, col_initial:col_final]

Indice de los renglones y las columnas

```python
dF.iloc[3:5,:5]
```
